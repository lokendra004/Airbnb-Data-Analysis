```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_Data.csv")
```

```python
print(df.isnull().sum())
```

```python
df.columns
```

## Handle missing values

```python
 # Convert 'last review' to datetime and handle errors
 df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
```

```python
 # Fill missing values
 df.fillna({'reviews per month': 0, 'last review': df['last review'].min()})
```

```python
# Drop records with missing 'name' or 'host name'
df.dropna(subset=['NAME', 'host name'])
```

```python
df=df.drop(columns=["house_rules","license"], errors='ignore')
```

```python
 # Remove dollar signs and convert to float
 df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
 df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

## Remove duplicates

```python
df.drop_duplicates()
```

## Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='red')  # Set histogram color
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

## Room Type Analysis

```python
 plt.figure(figsize=(8, 5))
 sns.countplot(x='room type', data=df , color='hotpink')
 plt.title('Room Type Distribution')
 plt.xlabel('Room Type')
 plt.ylabel('Count')
 plt.show()
```

##  Neighborhood Analysis

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood group', data=df,color="lightgreen" , order=df['neighbourhood group'].value_counts().index)
plt.title('Number of Listings by Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

## Relation blw price & room type

```python
 plt.figure(figsize=(10, 6))
 sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
 plt.title('Price vs. Room Type')
 plt.xlabel('Room Type')
 plt.ylabel('Price ($)')
 plt.legend(title='Room Type')
 plt.show()
```
